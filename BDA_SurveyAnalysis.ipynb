```python
# Framing thea dataset
df = pd.DataFrame(data)
```

```python
# Gender distribution
gender_distribution = df['Gender'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(gender_distribution, labels=gender_distribution.index, autopct='%1.1f%%', startangle=140,
        colors=['#66b3ff', '#ff9999'])
plt.title('Gender Distribution of Survey Participants')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python


# Count the occurrences of each selection method
selection_counts = df['Selection_Method'].value_counts()

# Plotting
plt.figure(figsize=(10, 6))
selection_counts.plot(kind='bar', color='skyblue')
plt.title('Selection Method for Watching Movies')
plt.xlabel('Selection Method')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Count the occurrences of each time of day
time_of_day_counts = df['Time_of_Day'].value_counts()

# Plotting
plt.figure(figsize=(10, 6))
time_of_day_counts.plot(kind='bar', color='green')
plt.title('Preferred Time of Day for Watching Movies')
plt.xlabel('Time of Day')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Count the occurrences of each frequency
frequency_counts = df['Frequency'].value_counts()

# Plotting
plt.figure(figsize=(10, 6))
frequency_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(len(frequency_counts))))
plt.title('Frequency of Movie Watching')
plt.ylabel('')  # Remove the y-label for a cleaner look
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
# Count the occurrences of each satisfaction level
satisfaction_counts = df['Satisfaction'].value_counts()

# Count the occurrences of each recommendation likelihood
recommendation_counts = df['Recommendation_Likelihood'].value_counts()

# Plotting Satisfaction
plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
satisfaction_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(len(satisfaction_counts))))
plt.title('Satisfaction')
plt.ylabel('')  # Remove the y-label for a cleaner look
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Plotting Recommendation Likelihood
plt.subplot(1, 2, 2)
recommendation_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(len(recommendation_counts))))
plt.title('Recommendation Likelihood')
plt.ylabel('')  # Remove the y-label for a cleaner look
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
# Plotting Participation in Discussions using a count plot
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Participation_in_Discussions', palette='pastel')
plt.title('Participation in Discussions')
plt.xlabel('Participation')
plt.ylabel('Count')
plt.show()
```
